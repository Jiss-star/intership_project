### Covert titles of Columns to Lower case

```python
def change_columns_to_lowercase(df):
    # Get the current column names
    old_column=df.columns.tolist()
    # Convert column names to lowercase
    new_col=[col.lower()for col in old_column]
    # Rename columns in the DataFrame
    df.columns = new_col
    return df


# Apply the function to change column names to lowercase

df=change_columns_to_lowercase(df)   
```

### Make the Space between the column names with underscore

```python
def replace_space_with_underscore(df):
    old_columns=df.columns.tolist()
    new_col=[col.replace(' ','_') for col in old_columns]
    df.columns=new_col
    return df

# Apply the function to replace spaces with underscores
df= replace_space_with_underscore(df)    
```

```python
def summary(df):
    print(f'data shape:{df.shape}')
    # Get data type
    dtype_df=pd.DataFrame(df.dtypes,columns=['Data Type'])
    ## Count missing values
    missing_values_df=pd.DataFrame(df.isnull().sum(),columns=['#missing'])
    missing_values_df['%missing'] = df.isnull().mean() * 100
    # Count unique values
    unique_values_df = pd.DataFrame(df.nunique(), columns=['#unique'])
    # Describe statistics
    describe_df = df.describe(include='all').transpose()
    stats_df = describe_df[['top', 'freq', 'min', 'max', 'mean', 'std']]
     # Count duplicate rows
    duplicate_rows = df.duplicated().sum()
    # Concatenate all dataframes
    summary_df =pd.concat([dtype_df, missing_values_df,unique_values_df,stats_df],axis=1)
    summary_df.loc['#duplicate'] = duplicate_rows
    styled_summary = summary_df.style.background_gradient(cmap='GnBu')
    return styled_summary

summary(df)
```

```python
df.columns
```

```python
#Dropping not needed variables
df1=df.drop(["unique_id","customer_remarks","order_id","order_date_time"],axis=1)
df1.sample(n=2)

#Replacing object dtype to timestamp dtype
timestamp_columns=['issue_reported_at','issue_responded','survey_response_date']
for col in timestamp_columns:
    df1[col]=pd.to_datetime(df1[col],dayfirst=True,errors='coerce')
    print(col,": dtype =",df1[col].dtype)
    
#In the context of the pd.to_datetime() function in Pandas, the dayfirst=True argument specifies the format in which the day comes before the month when parsing dates.
# When using pd.to_datetime(), some values in the data might not be valid dates or timestamps. By setting errors='coerce', 
#any invalid parsing will result in those entries being converted to
#NaT (Not a Time), which is pandas' way of representing missing or null values in datetime objects.
```

```python
#Replacing null value according to needs
for i in df1.columns:
# Impute missing values for categorical features-customer city and product category
    if (df1[i].isnull().sum()>0 and df1[i].dtype=="object"):
        df1[i].fillna("Unknown",inplace=True)
        print(i, "Variable has object dtype, so it got Replace its null values to Unknown \n")
## Impute missing values for numerical features        
    elif (df1[i].isnull().sum()>0 and df1[i].dtype=="float64"):
        df1[i].fillna(df1[i].median(),inplace=True)
        print(i, "Variable has float dtype, so it got Replace its null values to columns median value \n")
```
